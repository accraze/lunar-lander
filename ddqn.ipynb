```python
import gym, random, tempfile
import numpy as np
from gym import wrappers
import matplotlib.pyplot as plt

from model import build_model
from ddqn import DDQLAgent
```

```python
np.set_printoptions(precision=2)

env = gym.make('LunarLander-v2')
env = wrappers.Monitor(env, "./monitor")

num_states = env.observation_space.shape[0]
num_actions = env.action_space.n
```

```python
epsilon = 1
epsilon_decay = 0.9993
gamma = 0.99
lr = 0.0001

batch_size = 32
memory_limit=5000
```

```python
model = build_model(num_states, num_actions, lr)
target_model = build_model(num_states, num_actions, lr)
print(model.summary())
```

```python
agent = DDQLAgent(num_states, num_actions,
                  model=model, lr=lr, gamma=gamma,
                  epsilon=epsilon, epsilon_decay=epsilon_decay, 
                  memory_limit=memory_limit, env=env)

episodes = 3500
history = agent.train(episodes)
```

```python
all_rewards_train, epsilon_vals_train, reward_avg_train = history
```

```python
def plot(x, key=None, title='Rewards per episode', xlab='', ylab='', multi=False, save=True):
    if multi:
        for _ in x:
            plt.plot(_)
    else:
        plt.plot(x)
    plt.title(title)
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    if key is not None:
        plt.legend(key)
    
    plt.show()
    plt.close()
    
plot(all_rewards_train)
```

```python
alpha_vals = [0.0001] * len(epsilon_vals_train)
plt.plot(epsilon_vals_train, label='epsilon')
plt.plot(alpha_vals, label='alpha')
plt.title('Hyperparameters per episode')

plt.legend(loc='upper right')    
plt.show()
plt.close()
```

```python
# burn off the remaning steps of the episode
done = False
while not done:
    _,_,done,_ = env.step(0)
```

```python
# evaluate the algorithm for 100 episodes.
test_history = agent.test(100)
```

```python
plot(test_history[0],  title='Rewards per trial')
```
